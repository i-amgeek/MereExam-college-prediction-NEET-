```python
data1 = pd.read_csv('round_1_alloted.csv',header = None)
data1 = data1.set_index(0)
#data1.head()
```

```python
#data2.head()
```

```python
data1 = data1.drop(6,axis=1)
```

```python
data1.columns = ['Id', 'College','Course','6','7']
```

```python
#data1.head()
```

```python
data2.columns = data1.columns
```

```python
import re
def correct(data):
    data = re.sub(r'\n', ' ', str(data))
    return re.sub(r'\r', ' ', str(data))
    
```

```python
correct('jkshkfhsf\n jflksj\r sjlsj')
```

```python
out_data = data2.merge(data1, on='Id',how='outer')
out_data['College_x'] = out_data['College_x'].apply(correct)
out_data['College_y'] = out_data['College_y'].apply(correct)
#out_data.to_csv('out_data.csv')
#out_data['College_y']
```

```python
out_data.head()
```

```python
sum(out_data['College_x']=='nan')
out_data.tail()
```

```python
mask = (out_data['6_y'].isnull())
out_data.loc[mask,'6_y']=out_data.loc[mask,'6_x']
out_data.loc[mask,'7_y']=out_data.loc[mask,'7_x']
mask = (out_data['6_x'].isnull())
out_data.loc[mask,'6_x']=out_data.loc[mask,'6_y']
out_data.loc[mask,'7_x']=out_data.loc[mask,'7_y']
#out_data[mask].head()
```

```python
mask_col1 = out_data['College_x']=='nan'
out_data.loc[mask_col1, 'College_x'] = out_data.loc[mask_col1,'College_y']
#out_data[mask_col1].head()
mask_col2 = out_data['College_y']=='nan'
out_data.loc[mask_col2, 'College_y'] = out_data.loc[mask_col2,'College_x']
out_data[mask_col2].head()
```

```python
mask_course = out_data['Course_x'].isnull()
out_data.loc[mask_course,'Course_x']=out_data.loc[mask_course,'Course_y']
sum(out_data['Course_y'].isnull())
```

```python
out_data=out_data.drop(['College_y','Course_y','6_y','7_y'],axis=1)
out_data.head()
```
